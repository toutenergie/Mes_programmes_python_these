### Extraction des donnees de les Demande energetique

```python
def nettoyage_demande(dossier,fichiers_excel1,cellule):
    liste_demande = []
    for fichier in fichiers_excel1:
        
        chemin_complet = os.path.join(dossier, fichier) # pour avoir le chemin complet
        
        classeur = pd.read_excel(chemin_complet,sheet_name=None)
        
        for feuilles in classeur.keys():
            
            if feuilles == cellule: #'Demande':
        
                Demande = classeur[feuilles]

                Demande.columns  = classeur[feuilles].iloc[1]

                Demande = Demande[2:].reset_index(drop=True)[:27]

                Demande = Demande.loc[:,list(Demande.columns[:5])]
                liste_demande.append(Demande)
        Demande_2019 = pd.concat(liste_demande,axis=0)
    return Demande_2019 

Demande_janvier = nettoyage_demande(dossier,fichiers_excel1,'Demande')
Demande_janvier.head()
```

### Rapport des delestages

```python
def nottoyage_delestage(dossier,fichiers_excel1,cellule):
    liste_Délestage = []
    for fichier in fichiers_excel1:
        
        chemin_complet = os.path.join(dossier, fichier) # pour avoir le chemin complet
        
        classeur = pd.read_excel(chemin_complet,sheet_name=None)
        
        for feuilles in classeur.keys():
            
            if feuilles == cellule:
        
                Délestage = classeur[feuilles]

                Délestage.columns  = classeur[feuilles].iloc[1]
                
                Délestage =  Délestage[2:].reset_index(drop=True)[:27]
                
                Délestage = Délestage[Délestage.columns[:7]]
                
                liste_Délestage.append( Délestage) 
                
        Délestage = pd.concat(liste_Délestage,axis=0)
    return Délestage

delestage_janvier = nottoyage_delestage(dossier,fichiers_excel1,'Délestage')
```

### Extraction des donnees de les centrales

```python
def Nettoyage_centrale(dossier,fichiers_excel,cellule):
    liste_centrales = []
    for fichier in fichiers_excel:
        
        chemin_complet = os.path.join(dossier, fichier) # pour avoir le chemin complet
        
        classeur = pd.read_excel(chemin_complet,sheet_name=None) # Importation du classeur avec toutes les feuilles
        
        for feuilles in classeur.keys(): # iteration sur les feuilles du classeur
            
            if feuilles == cellule: # comparaison avec la feuille du nom de la cellule
        
                Centrale = classeur[feuilles] # recuperation du dataframe 

                Centrale.columns = Centrale.iloc[1] # Définir la première ligne restante comme en-têtes 

                Centrale = Centrale[2:].reset_index(drop=True)[:27] # Supprimer les deux premieres ligne apres l'en tetes
                
                liste_centrales.append(Centrale)
        
        # concatenation de l'ensemble des dataframe pour le mois de janvier        
        Centrales = pd.concat(liste_centrales,axis=0)
    return Centrales
```

### Extraction de la feuille Energie

```python
def nettoyages_energie(dossier,fichiers_excel1,cellule):
    liste_Pdispo = []
    for fichier in fichiers_excel1:
        
        chemin_complet = os.path.join(dossier, fichier) # pour avoir le chemin complet
        
        classeur = pd.read_excel(chemin_complet,sheet_name=None)
        
        for feuilles in classeur.keys():
            
            if feuilles == cellule:
        
                Pdispo = classeur[feuilles]
                
                Pdispo.columns = Pdispo.iloc[1]
                
                Pdispo  = Pdispo.reset_index(drop=True)[:27]
                
                liste_Pdispo.append( Pdispo) 
                
        Pdispo = pd.concat(liste_Pdispo,axis=0,ignore_index=True)
        
    return Pdispo
```
